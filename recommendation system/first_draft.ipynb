```python
def find_genres_keywords(x):
    final_list = []
    for i in ast.literal_eval(x):
        final_list.append(i['name'])
    return final_list

def find_cast(x):
    cast_list = []
    y = ast.literal_eval(x)
    max_iter = min(len(y),3)
    for i in range(0,max_iter):
        cast_list.append(y[i]['name'])
    return cast_list

def find_director(x):
    director_list = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            director_list.append(i['name'])
            break
    return director_list
```

```python
genre = [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
max(len(genre),3)
```

```python
df['genres']=df['genres'].apply(find_genres_keywords)
df['keywords'] = df['keywords'].apply(find_genres_keywords)
df['cast'] =  df['cast'].apply(find_cast)
df['crew']=df['crew'].apply(find_director)
```

```python
df['overview']=df['overview'].apply(lambda x: x.split())
```

```python
df['genres']=df['genres'].apply(lambda x: [i.replace(' ',"") for i in x])
df['keywords']=df['keywords'].apply(lambda x: [i.replace(' ',"") for i in x])
df['cast']=df['cast'].apply(lambda x: [i.replace(' ',"") for i in x])
df['crew']=df['crew'].apply(lambda x: [i.replace(' ',"") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df_new = df[['movie_id','title','tags']]
```

```python
df_new.head()
```

```python
df_new['tags']= df_new['tags'].apply(lambda x: ' '.join(x))
```

```python
nlp = spacy.load('en_core_web_md')
stop_words = nlp.Defaults.stop_words
```

```python
df['tags'][0]
```

```python
def data_cleaning_lemma(x):
    text = re.sub('[%s]' % re.escape(punc),'',x)
    doc = nlp(text)
    return ' '.join([tokens.lemma_.lower() for tokens in doc if tokens not in stop_words])
```

```python
df_new['tags'] = df_new['tags'].apply(data_cleaning_lemma)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000);
```

```python
word_vectors = cv.fit_transform(df['tags']).toarray()
```
